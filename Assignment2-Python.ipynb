```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sample_data/titanic_kaggle_traindata.csv")
df.head()
```

```python
df.info()
```

```python
#Question 1: Retrieve the details of the passenger in the 10th row

df.iloc[9:10,:]
```

```python
#Question 2: Extract the information of passengers from the 5th to the 15th row.

df.iloc[4:15,:]
```

```python
#Question 3: Retrieve the age and sex of passengers with PassengerId 100 to 150.

df.loc[99:149,['PassengerId','Age','Sex']]

#df[ (df['PassengerId']).between(100,150)].loc[:,['PassengerId','Age','Sex']]
```

```python
#Question 4: Extract the details of passengers from the 3rd to the 6th row and the 2nd to the 4th columns

df.iloc[2:6,1:4]
```

```python
#Question 5: Calculate the average fare paid by passengers

average_fare=df['Fare'].mean()
average_fare
```

```python
#Question 6: Group the data by 'Pclass' and calculate the median age for each passenger class.

median_age_by_pclass=df.groupby('Pclass')['Age'].median()
median_age_by_pclass.to_frame()
```

```python
#Question 7: Sort the DataFrame by 'Fare' in descending order.

df.sort_values('Fare',ascending=False)
```

```python
#Question 8: Retrieve the name and ticket number of passengers who survived.

df[df['Survived']==1][['Name','Ticket']]
```

```python
#Question 9:  Extract the subset of passengers who embarked from 'C' and paid a fare greater than $50.

df[ (df['Embarked']=='C') & ( df['Fare'] >50) ]
```

```python
#Question 10: Group the data by 'Embarked' and 'Sex', and calculate the average fare for each combination of embarkation point and gender.

df.groupby(['Embarked','Sex'])['Fare'].mean().to_frame()
```
