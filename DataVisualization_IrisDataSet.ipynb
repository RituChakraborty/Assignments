```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('/content/sample_data/Iris.csv')
iris.head()
```

```python
#1. What is the distribution of sepal lengths and widths among the Iris species?

sns.set_style('whitegrid')
fig,axes=plt.subplots()
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='muted', ax=axes)
axes.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)


#used scatterplot to show the Distribution regarding sepal length and petal width
#Inference  from plot :
#Setosa species generally have a lower sepal length but higher sepal width
#Virginia species generally have higher sepal length but lower sepal width
#Versicolor species generally have sepal length and width in between the above two species
```

```python
#2. How does the petal length vary across different Iris species?
fig,axes=plt.subplots(figsize=(5,5))
sns.stripplot(data=iris,x='Species', y='PetalLengthCm', color='purple', hue='Species', palette='muted' ,jitter=True)

#Inference from below plot to visualize different petal length different Iris species:
#Petal Lengths differ from one Iris species with another
#Setosa species generally have petal length varies from 1cm to 2cm (lowest petal length in comparison to two species)
#Versicolor species generally have petal length varies of 3.5cm to ~5.1cm (medium petal length in comparison to other two species)
#Verginica species generally have petal length varies from ~4.5 cm to ~6.9cm (highest petal length in comparison to other two species)
```

```python
#3. Is there a relationship between sepal length and petal length?
sns.lmplot(data=iris,x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='muted')

#used lmplot to show the relationship between sepal length and petal length
#Inference  from plot:
# We could infer from the below plots that SepalLength and PetalLength has positive linear correlation (i.e when one of them increases the other increases) for each species
```

```python
#4. Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?

iris_corr=iris.drop(['Id','Species'], axis=1).corr()
fig,axes=plt.subplots(figsize=(4,4))
sns.heatmap(data=iris_corr, annot=True, linewidths=2 )
sns.pairplot(iris.drop(['Id'], axis=1), hue='Species')

#Used heatmaps and pairplots to show pairwise relationship between features
#Inference  from plot:

#PetalLength and PetalWidth  have high positive correlation
#SepalLength and PetalLength have high positive correlation
#SepalLength and PetalWidth have high postive correlation

#SepalLength and SepalWidth have negative correlation
#SepalWidth and PetalWidth have negative correlation
#SepalWidth and PetalLength have negative correlation

#Pattern:  as PetalLength increases PetalWidth and SepalLength increases but SepalWidth decreases
```

```python
#5. Can we distinguish between Iris species based on sepal and petal measurements?

fig,axes=plt.subplots(ncols=2, figsize=(10,4))
sns.boxplot(data=iris,x='Species', y='SepalLengthCm', palette='muted',hue='Species',legend=False, ax=axes[0])
sns.boxplot(data=iris,x='Species', y='PetalLengthCm', palette='muted',hue='Species',legend=False,ax=axes[1])

#used boxplots to demonstrate sepal and petal measurements accross Iris species
#Inference from plot:
#We could infer that the sepal length and petal length features can be used to distinguish between Iris species, as there is a significant difference in the range of the values of these  features in each species.
#However, we see comparatively bigger overlaps in sepal length value among all three species, whereas petal lengths demonstrate a more  distinct  value range among all three species.
```
